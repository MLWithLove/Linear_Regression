# KHAI BÁO THƯ VIỆN

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from numpy.linalg import norm
from numpy.linalg import matrix_rank
from sklearn import datasets,linear_model
```

BỘ DỮ LIỆU DÙNG ĐỂ DỰ ĐOÁN GIÁ NHÀ

```python
DATA=pd.read_csv("../input/kc-house-data/kc_house_data.csv")
DATA
```

PHƯƠNG PHÁP ĐẠO HÀM, TÌM ĐIỂM CỰC TIỂU, DÙNG ĐẾN GIẢ NGHỊCH ĐẢO

```python
X = np.array(DATA[['sqft_living','sqft_lot','sqft_above','sqft_basement','sqft_living15','sqft_lot15','bathrooms']])
one=np.ones((X.shape[0],1))
Xbar=np.concatenate((one,X),axis=1)
Y= np.array(DATA['price'])
```

TÌM ĐIỂM CỰC TIỂU

```python
A=np.dot(Xbar.T,Xbar)
b=np.dot(Xbar.T,Y)
w=np.dot(np.linalg.pinv(A),b)
```

TÍNH L2 NORM CỦA LOSS FUNCTION KHI CHƯA CHIA CHO SỐ ĐIỂM DỮ LIỆU

```python
norm(np.dot(Xbar,w)-Y)
```

NGHIỆM

```python
print(w)
```

Kiểm tra tính singular

```python
print(matrix_rank(A))
```

PHƯƠNG PHÁP DÙNG THƯ VIỆN linear_model.LinearRegression()


```python
regr=linear_model.LinearRegression()
regr.fit(X,Y)
print(regr.intercept_)
```

TÌM L2 NORM CHO LOSS FUNCTION KHI CHƯA CHIA CHO SỐ ĐIỂM DỮ LIỆU

```python
print(norm(regr.predict(X)-Y))
```

NGHIỆM

```python
print(regr.intercept_)
print(regr.coef_)
```
